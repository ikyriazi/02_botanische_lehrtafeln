```python
xml_files = glob.glob('metadaten/*')
```

```python
ns = {'lido':'http://www.lido-schema.org'}
```

```python
for xml_file in xml_files:
    print(xml_file + ":")
    tree = ET.parse(xml_file)
    root = tree.getroot()
    resourceSets = root.findall(".//lido:resourceSet", ns)
    for resourceSet in resourceSets:
        resourseRepresentations = resourceSet.findall("./lido:resourceRepresentation[@{http://www.lido-schema.org}type='http://terminology.lido-schema.org/resourceRepresentation_type/preview_representation']", ns)
        for resourseRepresentation in resourseRepresentations:
            # get image uri
            image_uri = resourseRepresentation.find("./lido:linkResource", ns).text
            
            # change resolution
            image_uri = image_uri.replace('full/300,','full/800,')
            print(image_uri)
            
            # set name of image file
            # based on xml-parsing-singlefile script, the position of the second naniweb in the uri is 77
            # the position after "naniweb_" is 8
            image_name_start = 77 + 8
            # the end of the name has to be calculated backwards, because not all file names have the same number of digits
            # image_name_end = image_name_start + 14
            # image_name = image_uri[image_name_start:image_name_end+1]
            image_name = image_uri[image_name_start:-24]
            print(image_name)
            
            # download image
            persist_image('../../../Coding_da_Vinci/CdV_Niedersachsen_2020/Datengeber/Datensets/UniGöttingen - Zentrale Kustodie/Botanische Lehrtafeln der Uni Göttingen/Mediendateien',image_uri)
            # persist_image('test-bilder',image_uri)
            
    print("")
```
